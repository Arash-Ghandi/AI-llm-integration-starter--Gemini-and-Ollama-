### Setup & Authentication

environment. This section covers loading the required libraries, securely managing API keys, and performing an initial validation check on them.

Dependencies & Libraries
Four key libraries are utilized in this section:

- os: Used for interacting with the operating system and reading environment variables.

- dotenv: Used to securely load sensitive information (like API keys) from a .env file, preventing them from being hardcoded or exposed.

- IPython.display: Used to render and display text outputs beautifully using Markdown formatting within the notebook.

- openai: Since many AI providers (including Google Gemini) support OpenAI's standard client structure, we use this library to establish the connection.

Secure Environment Management
Sensitive credentials, such as GEMINI_API_KEY, are stored in a separate .env file. The load_dotenv(override=True) method loads these variables into the environment so our code can securely access them.

Initial API Key Validation
To prevent runtime errors, a simple conditional structure checks:

- Whether the API key is missing entirely.

- Whether the API key format is correct (valid Google Gemini keys typically start with AIz).

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

load_dotenv(override=True)
gimini_api_key = os.getenv('GEMINI_API_KEY')
gimini_base_url = os.getenv("GEMINI_BASE_URL")
gimini_model = os.getenv("GEMINI_MODEL")

if not gimini_api_key:
    print("No API key was found ")
elif not gimini_api_key.startswith("AIz"):
    print("An API key was found, but it doesn't start (AIz); please check you're using the right key")
else:
    print("API key found and looks good so far!")
```

### GIMINI
### Sending the First Prompt to the Model
- Defining the Prompt (text): A simple string variable containing our request to the model (in this case, "Tell me a fun fact").

- Initializing the Client (gemini): Creating an instance of the OpenAI class. Since Gemini supports an OpenAI-compatible API, we pass the base_url and api_key configured in the previous step.

- Creating the Completion (chat.completions.create): This method sends the payload to the server. The model parameter specifies which Gemini version to use, and the messages parameter defines the conversation structure (the user's role and the prompt content).

- Displaying the Output (display): The response is extracted from the API's standard JSON structure (response.choices[0].message.content) and beautifully rendered in the notebook using Markdown formatting.

```python
text="Tell me a fun fact"
gemini = OpenAI(base_url=gimini_base_url, api_key=gimini_api_key)
response = gemini.chat.completions.create(model=gimini_model, messages=[{"role":"user", "content": text}])
display(Markdown(response.choices[0].message.content))
```

### Connecting to Local LLMs via Ollama
This section demonstrates how to use the exact same code structure to interact with open-source models running locally on your own machine using Ollama.

Key Concepts
- OpenAI API Compatibility: Ollama runs a local server that mimics the OpenAI API structure. This means you don't need to rewrite your connection methods; you only need to update the base URL and model name.

Local Configurations:

- The ollama_base_url typically points to a local address (such as http://localhost:11434/v1).

- The api_key is not required by Ollama since the model runs locally, but a placeholder string like 'ollama' is passed to satisfy the OpenAI client requirement.

- Project Flexibility: This design allows developers to seamlessly switch their application backend between cloud-based models (like Gemini) and local models (like Llama 3 or Mistral) by simply updating environment variables.

```python
text="Tell me a fun fact"
ollama_base_url = os.getenv("OLLAMA_BASE_URL")
ollama_model = os.getenv("OLLAMA_MODEL")
ollama = OpenAI(base_url=ollama_base_url, api_key='ollama')
response = ollama.chat.completions.create(model=ollama_model, messages=[{"role": "user", "content": text}])
display(Markdown(response.choices[0].message.content))
```
